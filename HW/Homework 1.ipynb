### Section 5: Functions
For each of the following functions, pick a test input to verify your code works.
<ol>
    <li>Write a function <code>underscore(list)</code> that takes an input list of strings, <code>list</code>, and outputs one string with <code>_</code> between the previously separated elements. For example, <code>['this', 'is', 'a', 'list']</code> would output <code>'this_is_a_list'</code>.</li>
    <li>Write a function <code>factorial(n)</code> that takes input integer <code>n</code> and outputs $n!$ (recall $n! = n\cdot (n-1)\cdot (n-2)\cdots 3\cdot 2\cdot 1$) </li>
    <li>Write a function <code>fibonacci(n)</code> that takes input integer <code>n</code> and computes the $n$-th Fibonacci number (recall the next Fibonacci number in the sequence is defined by adding the two previous terms: 0,1,1,2,3,5,8,13,...)</li>
    <li>Write a function that converts integers to base 2. For example, the number 17 in base 2 is 10001 since $17 = 1\cdot 2^4+0\cdot 2^3+0\cdot 2^2+0\cdot 2^1+1\cdot 2^0$</li>
    <li>Write a function that converts a number into US standard money formatting (every 3 digits contain a comma and demical separates cents from dollar). That is, design a function that takes a number like $123456789.10$ and outputs $ \$123,456,789.10$ </li>
    <li>Write a function <code>reverse(seq)</code> that takes an input sequence <code>seq</code> and outputs the reversed version of the sequence.</li>
    <li>Write a function that adds up the elements in a given list.</li>
    <li>Write a function that finds the maximum in a list of integers (without using any built-in maximum finding functions)</li>
        </ol>

```python
#1
print('Part 1: ')
def underscore(lst):
    answer = lst[0]
    for word in lst[1:]:
        answer += '_' + word
    return answer
    
print(underscore(['I', 'am', 'in', 'MATH300']))

#2
print('Part 2: ')
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n-1)

print(factorial(4))

#3
print('Part 3: ')
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

print(fibonacci(5))

#4
print('Part 4: ')
def base2(n):
    answer = ''
    while n > 0:
        answer = str(n%2) + answer
        n = n // 2
    return answer

print(base2(15))

#5
print('Part 5: ')
def US_dollar(n):
    return '$' + str(n)

#This returns a dollar sign followed by the number
print(US_dollar(23242342.10))

#6
print('Part 6: ')
def reverse(seq):
    return seq[::-1]

print(reverse([5,7,2,3]))

#7
print('Part 7: ')
def add_element(lst):
    total = 0
    for num in lst:
        total += num
    return total

print(add_element([3,4,5,6,7]))

#8
print('Part 8: ')
def maximum(lst):
    biggest = lst[0]
    for num in lst:
        if num > biggest:
            biggest = num
    return biggest

print(maximum([5,7,2,6,4]))

```
