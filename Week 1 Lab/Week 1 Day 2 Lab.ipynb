```python
(2 ** 3) ** 2
```

Why are the answers different?

---


````python
print(2 ** 3 ** 2)
print((2 ** 3) ** 2)
print('The results are different python calculates double asterisks from right to left, and the parathesis changes the order slightly.')
print('The first one starts from the right, and calculates 3^2 which is 9. The next step is 2^9 which is equal to 512')
print('The second one calculates from the left in the parenthesis first, so 2^3 is equal to 8. Finally, 8^2 is 64')
````
