## Problem 3

A "leap year" is a calendar year where an extra day is added. The rules are as follows:
- If divisible by 400, then leap year
- If divisible by 100, then not a leap year
- If divisible by 4, then leap year
- Otherwise, not a leap year

Write a block of code that takes in the variable `year` and determines if it is a leap year or not.

````python
year = 2026

if year % 400 == 0:
    print('leap year')
elif year % 100 == 0:
    print('not a leap year')
elif year % 4 == 0:
    print('leap year')
else:
    print('not a leap year')
````

## Problem 5

Write a for-loop that keeps a list of the sum of the first `n` integers as `n` ranges from 1 to 30.

````python
sum = []
total = 0

for i in range (1,31):
    total += i
    sum.append(total)

print(sum)
    
````

## Problem 6
Write a for-loop that counts the number of even integers in a list. You can use the following list as an test:

```python
    
    numbers = [1,13,4,15,2,7,8]
```

````python
numbers = [1,13,4,15,2,7,8]
count = 0

for number in numbers:
    if number % 2 == 0:
        count += 1
print(count)
````

## Problem 7

This is Problem 1 of Project Euler found here: https://projecteuler.net/problem=1 . Project Euler is a collection of problems for mathematicians that can be good coding practice. I encourage you to check it out!

Give this problem a try:

"If we list all the natural numbers below 10 that are multiples of 3 or 5 we get 3,5,6, and 9. The sum of these multiples is 23.
Find the sum of all the multiples of 3 or 5 below 1000."

````python
total = 0

for number in range (1000):
    if number % 3 == 0 or number % 5 == 0:
        total += number

print(total)
    
````
