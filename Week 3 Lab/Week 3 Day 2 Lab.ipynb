### Problem 1
Plot the function $f(x)=x^2$ for $x\in [-1,1]$ using `np.linspace()` with 10, 100, 1000 slices. Make sure your plot is as informative as possible.

```python
import matplotlib.pyplot as plt
import numpy as np

x10 = np.linspace(-1,1,10)
y10 = x10**2

x100 = np.linspace(-1,1,100)
y100 = x100**2

x1000 = np.linspace(-1,1,1000)
y1000 = x1000**2

plt.plot(x10,y10, label = "10 points")

plt.plot(x100,y100, label = "100 points")

plt.plot(x1000,y1000, label = "1000 points")

plt.title("f(x) = x²")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()
```

### Problem 2
Pick a function $f(x)$ you like and plot some values using `plt.scatter()`. Make sure your plot is as informative as possible.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-20,20,100)
y = x**2 + 5*x - 3

plt.scatter(x,y, color = "purple")
plt.title("f(x) = x² + 5x - 3")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
