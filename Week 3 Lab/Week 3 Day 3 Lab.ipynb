1. Create your own dataframe with at least 3 columns and 4 rows.
2. Save the dataframe to a csv file.
3. Read the csv back in.
4. Plot the data.

````python
import pandas as pd
import matplotlib.pyplot as plt

data ={
    "age": [20, 18, 35, 18, 25],
    "name": ['Jaime', 'Robb', 'Cat', 'Jon', 'Josh'],
    "Major": ['CS', 'Biology', 'Physics', 'English', 'Statistics']
}

df = pd.DataFrame(data)

df.to_csv("dataframe.csv", index=False)
new_df = pd.read_csv("dataframe.csv")
print(new_df)

plt.bar(new_df["name"], new_df["age"])
plt.title("Student ages")
plt.xlabel("Name")
plt.ylabel("age")
plt.show()

````
