````python
plt.hist(df["Age"])
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

#it appears to be a normal distribution that is slightly skewed right
````

````python
df = df.drop_duplicates()

plt.boxplot(df["Age"])
plt.title("Age Box Plot")
plt.show()
````

````python
df["Sex"].value_counts().plot(kind="bar")
plt.title("Number of Passengers by Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()

#On the titanic, there were nearly twice as many men than women.
````
