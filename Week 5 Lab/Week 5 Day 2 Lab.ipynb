### Problem 1
Suppose $f(x)=x^2+6x+10$

1. Plot the function to visualize the minimum
2. Find the minimum using `scipy`

````python
#plotting the function
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 5, 500)
y = x**2 + 6*x + 10

plt.figure(figsize=(8,5))
plt.plot(x, y)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = x² + 6x + 10")

plt.show()
````

````python
#finding the minimum
from scipy.optimize import minimize

def f(x):
    return x**2 + 6*x + 10
result = minimize(f, x0=0)

print(result.x)
````


## Problem 2

Find the maximum of $50x-x^2$.

````python
x = np.linspace(0, 60, 500)

y = 50*x - x**2

plt.figure(figsize=(8,5))
plt.plot(x, y)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("f(x) = 50x - x²")

def objective(x):
    return -(50*x - x**2)

result = minimize(objective, x0=10)

print(result.x)

plt.show()
````
