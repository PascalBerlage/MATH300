### Problem 1
*Gradient descent* on a function $f(x)$ in the two dimensional case works as follows:
1. Start with an initial value $x_0$.
2. Compute $f'(x_0)$.
3. 
    - If $f'(x_0)>0$, move to $x_0-\epsilon$. 
    - If $f'(x_0)<0$, move to $x_0+\epsilon$. 
    - If $f'(x_0)=0$, stay put.
4. Repeat steps 1-3 for a certain amount of iterations.

Create a function that performs gradient descent. Be careful: your function should take in an iteration number so it does not run forever.

```python
def gradient_descent(df, x0, epsilon, iterations):
    x = x0

    for i in range (iterations):
        slope = df(x)

        if slope > 0:
            x = x - epsilon
        elif slope < 0:
            x = x + epsilon
        else:
            break

    return x
```

### Problem 2
Your function from above should be able to find minima. Use `%timeit` to compare the run time of your function and `minimize` from the `scipy.optimize` library. Rerun for multiple different starting points.

```python
from scipy.optimize import minimize

def f(x):
    return (x - 6)**2

def df(x):
    return 2 * (x - 6)

def gradient_descent(df, x0, epsilon, iterations):
    x = x0

    for i in range (iterations):
        slope = df(x)

        if slope > 0:
            x = x - epsilon
        elif slope < 0:
            x = x + epsilon
        else:
            break

    return x

%timeit gradient_descent(df, -10, 0.01, 5000)
%timeit minimize(f, -10)

%timeit gradient_descent(df, 0, 0.01, 5000)
%timeit minimize(f, 0)

#The results show that scipy.optimize.minimize ran faster than my gradient_descent function. My function took about 876–877 μs, while minimize took about 673–695 μs. This shows that SciPy's optimization is more efficient.
```
