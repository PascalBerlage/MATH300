### Problem 1
Suppose 

$$\frac{dy}{dt}=1.2y.$$

instead of .5 from lecture. Solve again.

How does the graph change?



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

def model(t, y):
    return 1.2 * y

y0 = [10]

t0 = 0
t1 = 10

solution = solve_ivp(
    model,
    [t0,t1],
    y0,
    dense_output=True
)

t = np.linspace(0,10,200)
y = solution.sol(t)[0]

plt.figure(figsize=(8,5))

plt.plot(t,y)

plt.xlabel("Time")
plt.ylabel("Population")

plt.title("Exponential Growth")

plt.grid()

plt.show()

#The graph is steeper and rises much faster because increasing the growth rate from 0.5 to 1.2 makes the population grow more rapidly.
```
